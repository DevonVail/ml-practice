```python
import torch
import torch.nn as nn
import math
# Assume we have some raw logits (outputs of the last fully connected layer)

logits = torch.tensor([[-0.2135, -0.0248, 3.9850]]) # High confidence in class #2
truth = [1, 0, 0] # Let's assume the truth is class #0 is correct

# Manual softmax
summation_of_exp = 0
softmax = []
for idx, logit in enumerate(logits[0]):
    summation_of_exp += math.exp(logit.item())

loss = 0
for idx, logit in enumerate(logits[0]):
    softmax.append(math.exp(logit.item())/summation_of_exp)
    loss += truth[idx]*math.log(softmax[idx])
loss *= -1

print(f"Manual Loss: {loss}")

nn_loss_fn = nn.CrossEntropyLoss()
truth_tensor = torch.tensor(truth, dtype=float)
truth_tensor = torch.reshape(truth_tensor, logits.shape)

nn_loss = nn_loss_fn(logits, truth_tensor)

print(f"NN Loss: {nn_loss}")
```
